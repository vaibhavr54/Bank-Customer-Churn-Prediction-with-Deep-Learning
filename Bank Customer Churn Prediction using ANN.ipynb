```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('bank_churn.csv')
```

```python
df
```

```python
df.shape
```

```python
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Geography'].unique()
```

```python
df['Gender'].unique()
```

```python
df1 = pd.get_dummies(df, ['Geography', 'Gender'], drop_first=True)
```

```python
df1
```

```python
df1.dtypes
```

```python
for column in df1.columns:
    df1[column] = df1[column].astype('float32')
```

```python
df1.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df2 = df1
```

```python
df2['CreditScore'] = scaler.fit_transform(df2[['CreditScore']])
```

```python
df2['Age'] = scaler.fit_transform(df2[['Age']])
```

```python
df2['Tenure'] = scaler.fit_transform(df2[['Tenure']])
```

```python
multi_col_scaled = scaler.fit_transform(df2[['Balance', 'NumOfProducts', 'EstimatedSalary']])
```

```python
multi_col_scaled_df = pd.DataFrame(multi_col_scaled, columns=['Balance', 'NumOfProducts', 'EstimatedSalary'])
```

```python
multi_col_scaled_df
```

```python
df3 = pd.concat([df2, multi_col_scaled_df], axis=1)
```

```python
df3
```

```python
df3.drop(['Balance', 'NumOfProducts', 'EstimatedSalary'], axis=1, inplace=True)
```

```python
df3.info()
```

```python
df3.describe()
```

```python
df3
```

```python
X = df3.drop('Exited', axis=1)
y = df3.iloc[:, 5:6]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(16, input_shape=(8,), activation='relu'),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(2, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
%%timeit -n1 -r1
with tf.device('/CPU:0'):
    model.fit(X_train, y_train, epochs=100)
```

```python
model2 = keras.Sequential([
    keras.layers.Dense(16, input_shape=(8,), activation='relu'),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(2, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
%%timeit -n1 -r1
with tf.device('/GPU:0'):
    model2.fit(X_train, y_train, epochs=100)
```

```python
model3 = keras.Sequential([
    keras.layers.Dense(16, input_shape=(8,), activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(2, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model3.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
%%timeit -n1 -r1
with tf.device('/CPU:0'):
    model3.fit(X_train, y_train, epochs=50)
```

```python
model4 = keras.Sequential([
    keras.layers.Dense(16, input_shape=(8,), activation='relu'),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(2, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model4.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
%%timeit -n1 -r1
with tf.device('/CPU:0'):
    model4.fit(X_train, y_train, epochs=60)
```

```python
model2.evaluate(X_test, y_test)
```

```python
y_pred = model2.predict(X_test)
```

```python
binary_pre = (y_pred > 0.5).astype(int)
binary_pre[0:5]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, binary_pre))
```

```python
print(confusion_matrix(y_test, binary_pre))
```

```python
# Working towards class imbalance
```

```python
y_test.value_counts()
```

```python
df_class_0 = df3[df3['Exited'] == 0]
df_class_1 = df3[df3['Exited'] == 1]
```

```python
df_class_1.shape
```

```python
df_class_0.shape
```

```python
class_zero_count, class_one_count = df3.Exited.value_counts()
```

```python
class_one_count
```

```python
class_zero_count
```

```python
sampled_class_zero = df_class_0.sample(class_one_count)
```

```python
sampled_class_zero.shape
```

```python
df_balanced_1 = pd.concat([sampled_class_zero, df_class_1], axis=0)
```

```python
df_balanced_1
```

```python
# New Training and testing splitting
```

```python
X = df_balanced_1.drop('Exited', axis=1)
y = df_balanced_1.Exited
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=421)
```

```python
X_train.shape
```

```python
model_x = keras.Sequential([
    keras.layers.Dense(16, input_shape=(8,), activation='relu'),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(2, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model_x.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model_x.fit(X_train, y_train, epochs=100)
```

```python
model_x.evaluate(X_test, y_test)
```

```python
y_x_pred = model_x.predict(X_test)
```

```python
y_x_pred
```

```python
y_x_p = (y_x_pred > 0.49).astype(int)
```

```python
y_x_p[:10]
```

```python
print(classification_report(y_test, y_x_p))
```

```python
#Method 2: Oversampling minority class
```

```python
oversampled_class_1 = df_class_1.sample(class_zero_count, replace=True)
```

```python
df_balanced2 = pd.concat([df_class_0, oversampled_class_1], axis=0)
```

```python
df_balanced2
```

```python
X = df_balanced2.drop('Exited', axis=1)
y = df_balanced2.Exited
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1, stratify = y)
```

```python
model_y = keras.Sequential([
    keras.layers.Dense(16, input_shape=(8,), activation='relu'),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(2, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model_y.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model_y.fit(X_train, y_train, epochs=100)
```

```python
model_y.evaluate(X_test, y_test)
```

```python
y_y_pred = model_y.predict(X_test)
```

```python
y_y_p = (y_y_pred > 0.49).astype(int)
```

```python
y_y_p[:5]
```

```python
print(classification_report(y_test, y_y_p))
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy='minority')
```

```python
X = df3.drop('Exited', axis=1)
y = df3.iloc[:, 5:6]
```

```python
y.value_counts()
```

```python
X_sm, y_sm = smote.fit_resample(X, y)
```

```python
y_sm.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.25, random_state=51)
```

```python
model_smote = keras.Sequential([
    keras.layers.Dense(16, input_shape=(8,), activation='relu'),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(2, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model_smote.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model_smote.fit(X_train, y_train, epochs=100)
```

```python
model_smote.evaluate(X_test, y_test)
```

```python
y_sm_pred = model_smote.predict(X_test)
```

```python
y_sm_p = (y_sm_pred > 0.49).astype(int)
```

```python
y_sm_p[:5]
```

```python
y_test[:5]
```

```python
print(classification_report(y_test, y_sm_p))
```
